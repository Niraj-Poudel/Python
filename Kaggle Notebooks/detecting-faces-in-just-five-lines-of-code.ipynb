```python
import PIL.Image
import PIL.ImageDraw
import face_recognition
from matplotlib.pyplot import imshow
```

### Prerequisites
* Pillow
     
     Python Imaging Library (abbreviated as PIL) is a free and open-source additional library for the Python programming language that adds support for opening, manipulating, and saving many different image file formats.
      
      
      
* Face Recognition

    Recognize and manipulate faces from Python or from the command line with
    the world’s simplest face recognition library.Built using dlib’s state-of-the-art face recognition built with deep learning. The model has an accuracy of 99.38% on the [Labeled Faces in the Wild](http://vis-www.cs.umass.edu/lfw/) benchmark.

----
## Line 1

In the first line of code I have used the face recognition library load_image_file() function to load the sample image from our respective directory. And using matplotlib imshow method to show the loaded image.

----

```python
input_image = face_recognition.load_image_file('/kaggle/input/billionaires.jpg')
imshow(input_image);
```


----
## Line 2

Passing our loaded image into face recognition library in the form of array. The face locations function in the face recognition library returns the total faces in the particular image location.

----

```python
image_faces = face_recognition.face_locations(input_image)
```

-----

## Line 3

Here our recognition library successfully detected the total number of faces in our sample image.Just a one line of above code is capable of detecting the appropriate faces. Look how powerful is [face recognition](https://pypi.org/project/face-recognition/) library.

-----

```python
print('Total number of faces in a image: ',len(image_faces))
```

----

## Line 4
Highlighting the faces in an image.

In this line i have used python image library to load the image and drawing the shape in the image

----

```python
output_image = PIL.Image.fromarray(input_image)
draw_shape = PIL.ImageDraw.Draw(output_image)
```

----

## Line 5

Getting each face locations and drawing rectangle in the each and every face in an image

----

```python
for faces in image_faces:
    top,left,bottom,right=faces
    draw_shape.rectangle([left,top,right,bottom],outline='red',width=20)
```

## Output

```python
imshow(output_image);
```

Here we successfully detect the face and draw a rectangle in each detected faces using python face recognition and PIL libraries.
